# 숫자형: 정수

```python
a = 123
print(a)
a = -178
print(a)
a = 0
print(a)
```

# 숫자형: 실수

```python
a = 1.2
print(a)
a = -3.48
print(a)
a = 4.24e10
print(a)
a = 4.24e-10
print(a)
```

# 숫자형: 8진수와 16진수

```python
a = 0o177
a
print(a)
a = 0x8FF
a
print(a)
a = 0xABC
a
print(a)
```

# 숫자형: 연산자와 연산

```python
a = 3
b = 4
print(a + b)
print(a / b)
print(a ** b)
print(a % b)
print(a // b)
```
